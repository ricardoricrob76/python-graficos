```python
# importando as biblitecas e classes do projeto.
import numpy as np  # trabalha com objetos vetor e matriz multidimensional
import matplotlib.pyplot as plt # gerar os gráficos de visualização.
```

```python
# Criando os campos Descritivos (Rótulos de dados)
labels = ["A1","A2","A3","A4","A5"]
```

```python
# Criando os valores dos campos (Métricas do Gráfico)
valores = [10, 34, 56, 78, 12]
```

```python
# Visualizando o gráfico com o objeto plt (Matplotlib)
plt.bar(labels,valores)
```

```python
# Visualizando o gráfico com o objeto plt (Matplotlib)
plt.barh(labels,valores)
```

```python
# Atribuindo novos aos eixos do gráficos (x, y)
plt.bar(labels,valores)
plt.xlabel('Código do Produto')
plt.ylabel('Valor do Produto')
```

```python
# Alterar as cores dos elementos gráficos.
plt.bar(labels,valores,color='silver')
plt.xlabel('ID')
plt.ylabel('Valor')
```

```python
# Colando os valores existentes em cada tipo como rótulo de dados.
plt.bar(labels,valores,color='red')
plt.xlabel('Código do Produto')
plt.ylabel('Valores dos Produtos')

for x,y in zip(labels,valores):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom')
```

```python
# Alterar as cores dos elementos gráficos.
plt.bar(labels,valores)
plt.xlabel('ID')
plt.ylabel('Valor')
```

```python
# Criando um novo Vetor chamado valores2.
valores2 = [8,30,60,21,30]
```

```python
# PLotar o gráfico com os dois gráficos e elementos de valores juntos.
largura = 0.4
pos1 = np.arange(len(valores))
pos2 = [x + largura for x in pos1]

plt.bar(pos1,valores, color='red', width=largura, label='Barra Valores 1')
plt.bar(pos2, valores2, color='blue', width=largura, label='Barra Valores 2')
plt.legend()
```

```python
plt.bar(pos1,valores,label = "Barra 1", color="pink")
plt.bar(pos1,valores2,label = "Barra 2", color="blue",bottom = valores)
```
